```python
# https://qiita.com/ikuzak/items/1332625192daab208e22

import sys

eps = 0.0               # 計算機イプシロン. 初期値は 0
tmp = 1.0               # tmpは計算機イプシロンを求める途中で使う変数. 初期値は 1

while 1.0 + tmp > 1.0:  # tmpがとみなされるまでは以下の計算を繰り返す
    eps = tmp           # まず、前のtmp値をepsilonに代入 
    tmp = tmp/2         # 次に、tmpを1/2 倍して小さい値へとアップデート

    print(eps,tmp)
print('final epsilon',eps)
print('sys.float_info.epsilon', sys.float_info.epsilon)
```
